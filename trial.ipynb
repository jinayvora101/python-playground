```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(
    data = {
        "markup": np.abs(np.random.standard_normal(1000)),
        "bill-code": np.uint(np.random.choice(np.linspace(1, 50, 50), size=1000)),
        "rucc": np.uint(np.random.choice(np.linspace(1, 9, 9), size=1000)),
    },
)
```

```python
df.groupby(["rucc", "bill-code"]).agg(
    start = ("markup", lambda k: k.mean()+k.std()),
    target = ("markup", lambda k: k.mean()),
    floor = ("markup", lambda k: k.mean()-k.std()),
).pivot_table(columns="rucc", index="bill-code", values=["start", "target", "floor"], aggfunc="sum").T.reset_index() \
 .assign(header = lambda k: k.level_0 + " " + k.rucc.astype(str)).sort_values(by="rucc") \
 .drop(["level_0", "rucc"], axis=1).set_index("header")#.reset_index()
```
