```python
standard_data_path = "C:\\Users\\Jinay Vora\\Documents\\Spotify_Data_\\Spotify_Standard_Data"
chdir(standard_data_path)

file_list = listdir()
del file_list[[i.find(".pdf")>=0 for i in file_list].index(True)]
file_list = {file_list[i]: i for i in range(len(file_list))}

json_list = []
for i in file_list:
    with open(standard_data_path + "\\" + i, "r", encoding="utf8", errors="ignore") as f:
        json_list.append(json.loads(f.read()))
del f, i, standard_data_path
```

```python
streaming_data = pd.DataFrame.from_dict(json_list[file_list["StreamingHistory0.json"]])
streaming_data["endTime"] = [dt.strptime(i, "%Y-%m-%d %H:%M") for i in streaming_data["endTime"].values]

def summarise_streaming_data(df, pivot_col):
    time = pd.pivot_table(df, index=pivot_col, values="msPlayed", aggfunc=np.sum).sort_values(by="msPlayed", ascending=False).reset_index()
    time["msPlayed"] = time["msPlayed"]/(60*1000)
    time.rename(columns={"msPlayed": "minPlayed"}, inplace=True)
    count = pd.pivot_table(df, index=pivot_col, values="msPlayed", aggfunc=np.size).sort_values(by="msPlayed", ascending=False).reset_index()
    count.rename(columns={"msPlayed": "count"}, inplace=True)
    return time.merge(count, how="outer")

streaming_data_artist_pivot = summarise_streaming_data(streaming_data, "artistName")
streaming_data_track_pivot = summarise_streaming_data(streaming_data, "trackName")

# streaming_data['a'] = np.random.rand(8965)*10000
# streaming_data['b'] = np.random.rand(8965)*10000
# streaming_data['c'] = np.random.rand(8965)*10000
# streaming_data['d'] = np.random.rand(8965)*10000
pd.set_option('display.expand_frame_repr', False)
print(streaming_data)
```
